```python
countries = results.find_all("div", class_="col-md-4 country")
# countries
```

```python
for country in countries:
    print(country)
    # print(country.prettify())
    break
```

```python
country_name = country.find('h3', class_='country-name')

country_text = country_name.get_text(strip=True) #  removes extra whitespace
print(country_text)
```

```python
country_info = country.find(class_='country-info')

country_capital = country_info.find(class_='country-capital').get_text(strip=True)
country_population = country_info.find(class_='country-population').get_text(strip=True)
country_area = country_info.find(class_='country-area').get_text(strip=True)
print(country_capital)
print(country_population)
print(country_area)
```

```python
# Creating a DataFrame

import pandas as pd
```

```python
countries_data = []  # Initialize a list to store country dictionaries

for country in countries:
    country_name       = country.find('h3', class_='country-name').get_text(strip=True)
    
    country_info       = country.find(class_='country-info')
    
    country_capital    = country_info.find(class_='country-capital').get_text(strip=True)
    country_population = country_info.find(class_='country-population').get_text(strip=True)
    country_area       = country_info.find(class_='country-area').get_text(strip=True)
    
    countries_data.append({
            'country_name': country_name,
            'country_capital': country_capital,
            'country_population': country_population,
            'country_area': country_area
        })
    
countries_df = pd.DataFrame(countries_data)
```

## Second part: Sustainalytics 

```python
URL = 'https://www.sustainalytics.com/esg-rating/'
stock = 'apple-inc/1007903183'

page = requests.get(URL+stock)
soup = BeautifulSoup(page.content, 'html.parser')
results = soup.find(class_='col-md-8 col-sm-12 company-data')
# results
```

```python
company_name = results.find(class_='row company-name').\
    find(class_='col').\
    get_text(strip=True)
```

```python
industry_group = results.find(class_='row intro-info').\
    find(class_='industry-group').\
    get_text(strip=True)
```

```python
country = results.find(class_='row intro-info').\
    find(class_='country').\
    get_text(strip=True)
```

```python
print(company_name)
print(industry_group)
print(country)
```

```python
results.find_all(class_='row intro-info')
```

```python
results = soup.find(class_='competitive-comparison')
table = results.find(class_='table comparison-table')
table_body = table.find('tbody')
# table_body
```

```python
rows = table_body.find_all('tr')

# Extract data from each row
for row in rows:
    # Extract company name
    company_tag = row.find('td', class_='cc-company-name').a
    company_name = company_tag.text.strip()
    company_href = company_tag['href'].strip() 
    
    # Extract risk rating value
    risk_rating = row.find('div', class_='cc-crr-item col-md-3').strong.text.strip()
    
    # Extract risk category
    risk_category = row.find('div', class_='cc-crr-item col-md-4').strong.text.strip()
    
    # Extract industry rank and total
    industry_rank = row.find('span', class_='cc-company-industry-group-rank').text.strip()
    industry_total = row.find('span', class_='cc-company-industry-group-positions-total').text.strip()
    
    # Print extracted data
    print(f"""
    Company: {company_name}, ({URL + company_href[3:]})
    Risk Rating: {risk_rating}, 
    Risk Category: {risk_category}, 
    Rank: {industry_rank} out of {industry_total}""")
```
